```python
#Basic Lineplot
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10, 6))
```

```python
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y, label='sin(x)')
```

```python
plt.title('Basic Line Plot')
```

```python
plt.xlabel('x')
```

```python
plt.ylabel('y')
```

```python
plt.legend()
```

```python
plt.savefig('line_plots.png')
```

```python
plt.close()
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.subplot(1, 2, 1)
```

```python
plt.plot(x, np.sin(x), 'b-', label='sin(x)')
```

```python
plt.title('sine')
```

```python
plt.legend()
```

```python
plt.subplot(1, 2, 2)
```

```python
plt.plot(x, np.cos(x), '-r', label='cos(x)')
```

```python
plt.title('cosine')
```

```python
plt.legend()
```

```python
plt.savefig('multiplots.png')
```

```python
plt.close()
```
