```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100)+2
})
```

```python
plt.figure(figsize=(8, 6))
```

```python
df.plot(kind='hist', alpha=0.5)
```

```python
plt.title('Histogram')
```

```python
plt.savefig('histogram.png')
```

```python
plt.close()
```
