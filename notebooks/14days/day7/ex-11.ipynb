```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("\n === Scatterplot ===")
```

```python
plt.figure(figsize=(8, 6))
```

```python
x = np.random.rand(50)
```

```python
y = np.random.rand(50)
```

```python
colors = np.random.rand(50)
```

```python
plt.scatter(x, y, c=colors, alpha=0.5)
```

```python
plt.title('Scatter_plot')
```

```python
plt.savefig('scatter_plot.png')
```

```python
plt.close()
```
